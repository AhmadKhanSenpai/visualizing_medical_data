```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv("medical_examination.csv")
df.head()
```

```python
# 2
df["overweight"] = (df["weight"] / ((df["height"] / 100) ** 2) > 25).astype(int)
df.head()
```

```python
# 3
df["cholesterol"] = (df["cholesterol"] > 1).astype(int)
df["gluc"] = (df["gluc"] > 1).astype(int)
df.head()
```

```python
# 4
df_cat = df.melt(
    id_vars=["cardio"],
    value_vars=["cholesterol", "gluc", "smoke", "alco", "active", "overweight"],
    var_name="feature",
    value_name="value",
)
df_cat
```

```python
# 5
df_grouped = (
    df_cat.groupby(["cardio", "feature", "value"]).size().reset_index(name="total")
)
df_grouped
```

```python
sns.catplot(
    data=df_grouped,
    x="feature",
    y="total",
    hue="value",
    kind="bar",
    col="cardio",
)

plt.show()
```

```python
df_heat = df[
    (df["ap_lo"] <= df["ap_hi"])
    & (df["height"] >= df["height"].quantile(0.025))
    & (df["height"] <= df["height"].quantile(0.975))
    & (df["weight"] >= df["weight"].quantile(0.025))
    & (df["weight"] <= df["weight"].quantile(0.975))
]
df_heat.head()
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones(corr.shape, dtype=bool))
mask
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
```

```python
sns.heatmap(
    corr,
    mask=mask,
    square=True,
    linewidths=0.5,
    annot=True,
    fmt="0.1f",
    ax=ax,
    cmap="coolwarm",
)
```

# Heat Map Analysis

The heat map visualizes the correlation matrix, showing the relationships between various features in the dataset. Each cell represents the correlation coefficient between two variables, with values ranging from **-1** (strong negative correlation) to **1** (strong positive correlation). Below are key observations:

## Key Observations:

1. **Strong Positive Correlation**:

   - `weight` and `overweight` have a high positive correlation (**0.7**), indicating that as weight increases, the likelihood of being classified as overweight also rises significantly.

2. **Moderate Correlation**:

   - `gluc` (glucose levels) and `cholesterol` show a moderate positive correlation (**0.4**), suggesting that these two health metrics may rise together.
   - `weight` and `height` exhibit a weak-to-moderate correlation (**0.5**), which reflects that taller individuals may tend to weigh more.

3. **Weak Correlation**:

   - `cholesterol` and `cardio` have a weak positive correlation (**0.2**), meaning high cholesterol levels may slightly increase the risk of cardiovascular diseases.
   - `age` and `cardio` have a weak positive correlation (**0.2**), suggesting that cardiovascular disease is more likely as age increases.

4. **Negligible Correlations**:

   - Features like `smoke`, `alco`, and `active` show very low correlations (close to **0**) with `cardio`. These relationships may either be weak or non-linear.

5. **Negative Correlations**:
   - `height` and `overweight` have a weak negative correlation (**-0.1**), meaning taller individuals are slightly less likely to be classified as overweight.

## Conclusion:

The heat map highlights significant positive correlations such as `weight` with `overweight` and `gluc` with `cholesterol`. Weak or negligible correlations, such as those for lifestyle factors (`smoke`, `alco`, and `active`) with `cardio`, suggest limited linear relationships and may warrant further exploration with non-linear models or additional data.
